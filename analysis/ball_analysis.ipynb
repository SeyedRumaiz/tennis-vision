```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the pickle file
with open('../tracker_stubs/ball_detections.pkl', 'rb') as f:
    ball_positions = pickle.load(f)
```

```python
ball_positions = [x.get(1,[]) for x in ball_positions]
# convert the list into pandas dataframe
df_ball_positions = pd.DataFrame(ball_positions,columns=['x1','y1','x2','y2'])

# interpolate the missing values
df_ball_positions = df_ball_positions.interpolate()
df_ball_positions = df_ball_positions.bfill()
```

```python
df_ball_positions['mid_y'] = (df_ball_positions['y1'] + df_ball_positions['y2'])/2
df_ball_positions['mid_y_rolling_mean'] = df_ball_positions['mid_y'].rolling(window=5, min_periods=1, center=False).mean()
```

```python
# Ball speed per frame
df_ball_positions["center_x"] = (df_ball_positions["x1"] + df_ball_positions["x2"]) / 2
df_ball_positions["center_y"] = (df_ball_positions["y1"] + df_ball_positions["y2"]) / 2

df_ball_positions["dx"] = df_ball_positions["center_x"].diff()
df_ball_positions["dy"] = df_ball_positions["center_y"].diff()

df_ball_positions["pixel_speed"] = (
    df_ball_positions["dx"]**2 +
    df_ball_positions["dy"]**2
) ** 0.5
```

```python
# Acceleration
df_ball_positions["acceleration"] = (
    df_ball_positions["pixel_speed"].diff()
)
```

```python
# Ball speed over time
plt.figure(figsize=(15,5))
plt.plot(df_ball_positions["pixel_speed"])
plt.title("Ball Speed Over Time")
plt.show()
```

```python
# Acceleration over time
plt.figure(figsize=(15,5))
plt.plot(df_ball_positions["acceleration"])
plt.title("Ball Acceleration Over Time")
plt.show()
```

```python
# plot mid_y_rolling_mean
plt.plot(df_ball_positions['mid_y_rolling_mean'])
plt.title('Ball Mid Y Rolling Mean')
plt.show()
```

```python
df_ball_positions['delta_y'] = df_ball_positions['mid_y_rolling_mean'].diff()
```

```python
# plot mid_y_rolling_mean
plt.plot(df_ball_positions['delta_y'])
plt.title('Ball Mid Y Rolling Mean Delta')
plt.show()
```

```python
df_ball_positions['ball_hit']=0
```

```python
minimum_change_frames_for_hit = 25
for i in range(1,len(df_ball_positions)- int(minimum_change_frames_for_hit*1.2) ):
    negative_position_change = df_ball_positions['delta_y'].iloc[i] >0 and df_ball_positions['delta_y'].iloc[i+1] <0
    positive_position_change = df_ball_positions['delta_y'].iloc[i] <0 and df_ball_positions['delta_y'].iloc[i+1] >0

    if negative_position_change or positive_position_change:
        change_count = 0 
        for change_frame in range(i+1, i+int(minimum_change_frames_for_hit*1.2)+1):
            negative_position_change_following_frame = df_ball_positions['delta_y'].iloc[i] >0 and df_ball_positions['delta_y'].iloc[change_frame] <0
            positive_position_change_following_frame = df_ball_positions['delta_y'].iloc[i] <0 and df_ball_positions['delta_y'].iloc[change_frame] >0

            if negative_position_change and negative_position_change_following_frame:
                change_count+=1
            elif positive_position_change and positive_position_change_following_frame:
                change_count+=1
    
        if change_count>minimum_change_frames_for_hit-1:
            df_ball_positions['ball_hit'].iloc[i] = 1

frame_nums_with_ball_hits = df_ball_positions[df_ball_positions['ball_hit']==1].index.tolist()
```
